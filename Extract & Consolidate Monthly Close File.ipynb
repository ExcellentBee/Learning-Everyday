## II. Use openpyxl

```python
import glob, os, openpyxl
from openpyxl.utils import get_column_letter

# Get directory and file name
os.chdir("C:/Users/Eliza Bui/Automation Examples/Extract and Consolidate Monthly Close Files")

found_parent_file = None  # Initialize a variable to track the found file

for file in glob.glob("*RevRec Snapshot ASC606*.xlsx"):
    if "Parent" in file:
        found_parent_file = file
        print(f"Found the file: {found_parent_file}")
        break

if found_parent_file:
    # Open the RevRec file
    wb_parent = openpyxl.load_workbook(found_parent_file)

    # Create a new tab to copy data over
    ws2 = wb_parent.create_sheet('New DataCut')
    ws1 = wb_parent.active  # Get the active worksheet

    # Create a list of specific columns in perferred order
    Report1_Column_List = ['RR Number', 'Performance Obligation Line Item Number', 'Account Formal Name', 'Account ID', 
'Counter ID', 'Subscription #', 'Contract Number', 'Program Code', 'Product: Product Name', 'Product: Product Code',
'Parent Product: Product Code', 'Parent Product: Product Name', 'Product: Product Family', 'Fee Type', 'Risk Status', 
'Risk Type', 'Recognition Start Date', 'Recognition End Date', 'Start Date', 'End Date', 'Revenue', 'GAAP Amount',
'Recognized to Date',  'Deferred Amount', 'Subscription Type', 'Line Status', 'Deployment ID', 'Deployment Name',
'Comments', 'Contract Term Type', 'Subscription Type', 'Revenue Recognition Template: Template Name','Contracted Quantity',                            
'Actual Quantity', 'Contracted Hours', 'Actual Hours', 'Performance Obligation Complete', 'Cost Center' ]

    # Loop through all column names, look for the one matching indexed element of Column List
    for i in range (0, len(Report1_Column_List)):
        for j in range(1, ws1.max_column):
            if ws1.cell(row = 1, column = j).value == Report1_Column_List[i]:
                Col_Letter = get_column_letter(j)
        Current_Col = ws1[Col_Letter +'1:' + Col_Letter + str(ws1.max_row)]

        for row_id, tuple_cell in enumerate(Current_Col, 1):
            for cell in tuple_cell:
                ws2.cell(row = row_id, column = i+1 ).value = cell.value

    # Save and close the workbook
    wb_parent.save(found_parent_file)
    wb_parent.close()
    print("Parent RevRec snapshot extraction DONE!")

else:
    print("File not Found")
```
